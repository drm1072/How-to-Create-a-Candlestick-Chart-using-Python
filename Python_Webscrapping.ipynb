**Article**

```python
from newspaper import Article #using this website to scrape

url = 'https://towardsdatascience.com/10-things-i-wish-i-knew-about-learning-data-science-7a30bfb91759'
article = Article(url) #article variable
article.download()
```

```python
print(article.html)
```

**Parse Article**

```python
article.parse()
```

```python
article.authors
```

```python
article.publish_date
```

```python
print(article.text) #prints entire article contents
```

```python
print(article.top_image) #code to show top image
```

**NLP- Natural Language Processing**

```python
import nltk #natural language processing toolkit for python
nltk.download('punkt') #punkt tokenizer

article.nlp()
```

```python
article.keywords #keywords within article
```

```python
print(article.summary) #summarizes artilce
```
