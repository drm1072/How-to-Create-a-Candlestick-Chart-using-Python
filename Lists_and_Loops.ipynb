**A while loop** is a loop that is repeated as long as an expression is true.  An expression is a statement that has a value.

A **do while loop** or repeat until loop repeats until an expression becomes fate.

An **infinite or endless loop** is a loop that repeats indefintely

A **nested loop** appears inside any other for loop, while loop, or do loop.

```python
#A "while loop" starts from the start of the variable and ends until our desire is over 
A = 16
while A < 40:
  print(A)
  A += 2
```

```python
A = 16
while A > 40 :
   print(A)
   A += 2
```

```python
A = 60 #As you can see this will run until eternity!!!
while A > 40 :
  print(A)
  A += 2
```
